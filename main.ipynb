```python
# membaca file excel menjadi pandas dataframe
df_country = pd.read_excel("dataset/Country-Code.xlsx")
df_country.head()
```

```python
# menggabungkan data dengan merge, menggunakan metode left join
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df
```

```python
# cara lain cek tipe data
final_df.dtypes
```

```python
# melihat rekord data dari setiap negara, top 3 pengguna zomato
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
# country_val
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.2f%%')
```

## Hasil Pengamatan
> Rekord maksimal transaksi zomato berasal dari india disusul oleh USA dan UK

```python
# menggabungkan data menggunakan group by
ratings = final_df.groupby(["Aggregate rating", "Rating color", 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## Hasil Pengamatan
1. Saat rating range diantara 4.5-4.9 = Excellent
2. Saat rating range diantara 4.0-4.4 = Very Good
3. Saat rating range diantara 3.5-3.9 = Good
4. Saat rating range diantara 3.0-3.4 = Average
5. Saat rating range diantara 2.5-2.9 = Average
4. Saat rating range diantara 2.0-2.4 = Poor


```python
# mengubah besar canvas
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",hue="Rating color", data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'lime', 'green'])
```

## Hasil Pengamatan
1. Not rated (user tidak memberikan penilaian) data terhitung sangat tinggi
2. Jumlah rating maksimal berada diantara 2.5 - 3.4

```python
## menghitung total rekord
sns.countplot(x="Rating color", data = ratings, palette=['blue', 'red', 'orange', 'yellow', 'lime', 'green'])
# ratings
```
